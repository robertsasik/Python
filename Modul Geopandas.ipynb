# Čo je Geopandas

1. GeoPandas je projekt s otvoreným zdrojovým kódom, ktorý uľahčuje prácu s geopriestorovými údajmi v programovacom jazyku Python. 
2. GeoPandas rozširuje dátové typy používané v module Pandas, aby umožnili priestorové operácie s geometrickými dátovými typmi. 
3. Geometrické operácie sa vykonávajú pomocou knižnice shapely. 
4. Geopandas ďalej závisí od knižnice Fiona pre prístup k súborom a od Matplotlib pre vykresľovanie.

## Vizualizácia vektorových dát

```python
#Vykreslanie vektorovej vrstvy
gdf.plot()
```

```python
#Vykreslanie vektorovej vrstvy podľa atribútov (napr. podľa krajov - NM2).
gdf.plot("NM2")
```

```python
#Vykreslanie vektorovej vrstvy podľa atribútov a zobrazenie legendy.
gdf.plot("NM2", legend = True)
```

```python
#Zobrazenie a editácia legendy.
import matplotlib.pyplot as plt
fig, ax = plt.subplots (1, figsize = (7, 10))
gdf.plot (ax=ax, column = "NM2", legend=True, legend_kwds={"loc": "center left"})
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.05, 0.5))
```
